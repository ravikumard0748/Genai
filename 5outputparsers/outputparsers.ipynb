```python
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_groq import ChatGroq
```

```python
# without using any output parser
prompt = PromptTemplate(
    input_variables = ["Theme"],
    template = "Give the one obest book in this Theme : {Theme}"
    )
llm = ChatGroq(model = "openai/gpt-oss-120b")
chain = prompt | llm 
```

```python
output = chain.invoke({"Theme" : "Datascience"})
```

```python
print(output)
```

```python
from langchain_core.output_parsers import StrOutputParser
chain = prompt | llm | StrOutputParser()
```

```python
output = chain.invoke({"Theme" : "Datascience"})
```

```python
print(output)
```

```python
from langchain_core.output_parsers import CommaSeparatedListOutputParser
output_parser=CommaSeparatedListOutputParser()
output_format = output_parser.get_format_instructions()
output_format
prompt = PromptTemplate(input_variables = ["Theme"],
    template = "Give the top 5 best book with their author name  in this Theme : {Theme} \n {output}"
    )
llm = ChatGroq(model = "openai/gpt-oss-120b")
chain = prompt | llm | output_parser
output = chain.invoke({"Theme" : "Datascience","output" : output_format})
```

```python
print(output)
```
