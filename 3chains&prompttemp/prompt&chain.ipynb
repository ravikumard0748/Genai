```python
from langchain_core.prompts import PromptTemplate
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_classic import LLMChain
from langchain_classic.chains import SequentialChain
```

```python
llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    temperature = 0.5,
)
```

```python
book_name_prompt = PromptTemplate(
    input_variables=["Theme"],
    template = """
    Please provide a simple list of ten well-known books that center around the theme of
    Theme of : {Theme}.
    Do not include book description 
    """ 
)
```

```python
book_name_chain = LLMChain(llm=llm,prompt=book_name_prompt,output_key="book_names_list")
```

```python
book_list = book_name_chain.invoke(input="Datascience")
```

```python
print(book_list["book_names_list"])
```

```python
book_summary_prompt_template = PromptTemplate(
    input = ["book_names_list"],
    template = """
    Give the all books with its author name only and then
    Take one book from the books list {book_names_list}.
    Provide a comprehensive summary of the book,in three sections
    and each section with three summary points """
)
```

```python
book_summary_chain = LLMChain(
    llm = llm , prompt = book_summary_prompt_template,
    output_key = "book_summary"
)
```

```python
from langchain_classic.chains import SequentialChain
```

```python
book_chain = SequentialChain(
    chains =[book_name_chain,book_summary_chain],
    input_variables = ["Theme"],
    output_variables = ["book_names_list","book_summary"]  
)
```

```python
result = book_chain.invoke({"Theme":"Personal Finance"})
```

```python
print(result["book_summary"])
```

```python
from langchain_core.prompts import PromptTemplate
from langchain_google_genai import ChatGoogleGenerativeAI
```

```python
from langchain_core.output_parsers import StrOutputParser
analyzing_prompt = PromptTemplate(
    input_variables = ["stock_details"],
    template = """
    Analyze the details of the given stock and tell the Top 3 positive and negative side of this stock :
    {stock_details}
    """,
)

llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash")
chain1 = analyzing_prompt | llm | StrOutputParser()
```

```python
from langchain_core.runnables import RunnablePassthrough
main_chain  = chain1 |(lambda x : {"analysis" : x} )| chain2
```

```python
main_chain.invoke({"stock_details" : stock_details})
```

```python
from langchain_core.prompts import ChatPromptTemplate,MessagesPlaceholder
from langchain_core.messages import HumanMessage,AIMessage
from langchain_groq import ChatGroq
chat_history = []
prompt = ChatPromptTemplate([
    ("system","You are helpful assistant"),
    MessagesPlaceholder(variable_name = "chat_history"),
    ("human","{question}")
])
llm = ChatGroq(model = "openai/gpt-oss-120b")

chain = prompt | llm
output = chain.invoke({"question" : "Are you know about the book Power Of habit","chat_history":[]})
```

```python
chat_history.append(AIMessage(content = "Are you know about the book Power Of habit"))
chat_history.append(HumanMessage(content = output.content))
```

```python
second_chat = chain.invoke({"question" : "Summarize the above book core meaning in one line","chat_history":chat_history})
```

```python
second_chat.content
```

```python
#  few Shot prompts 
from langchain_core.prompts import FewShotPromptTemplate,PromptTemplate

examples = [
    {"input" :"This is amazing movie" ,"output":"positive"},
    {"input" : "It is waste of money" , "output" : "negative"}
]

prompt = PromptTemplate(
    input_variables = ["input","output"],
    template = "Input {input} /n output{output}"
)

main_template = FewShotPromptTemplate(
    prefix = "Classify the sentences based on their sentiment",
    examples = examples,
    example_prompt = prompt,
    suffix = "Input : {sentence} /n Output : ",
    input_variables = ["sentences"]
    
)

```

```python
chain = main_template | llm
```

```python
output=chain.invoke({"sentence":"I am living with my family because it is my fate because i don't love any one"})
```
