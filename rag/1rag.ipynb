```python
from langchain_community.document_loaders import PyPDFLoader
loader = PyPDFLoader("text_processing.pdf")
```

```python
document = loader.load()
```

```python
final_document=""
for page in document:
    final_document+=page.page_content
```

```python
print(final_document)
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(chunk_size = 300,chunk_overlap = 30)
```

```python
chunks = text_splitter.split_documents(document)
```

```python
print(chunks)
```

```python
from langchain_huggingface import HuggingFaceEndpointEmbeddings

embeddings = HuggingFaceEndpointEmbeddings(
    model="sentence-transformers/all-MiniLM-L6-v2",
)

```

```python
from langchain_community.vectorstores import Chroma,FAISS,Qdrant,Pinecone
```

```python
text_db = Chroma.from_documents(chunks,embeddings,persist_directory="text_db")
```

```python
retriver = text_db.as_retriever()
```

```python
result = retriver.invoke("What is lemmatization")
```

```python
print(result)
```
