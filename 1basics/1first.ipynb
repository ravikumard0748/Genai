```python
from langchain_google_genai import ChatGoogleGenerativeAI
llm = ChatGoogleGenerativeAI(
    model = "gemini-2.5-flash",
    temperature = 0.7
)
```

```python
llm.invoke("what are the things i need to know for an generative ai engieer")
```

```python
result = llm.invoke("how many hours a person must sleep in a day")
```
